```python
# I want to know that virat kohli ne kis team ke against sabse jayada runs mare
vk = delivery[delivery['batsman'] == 'V Kohli']
```

```python
vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head()
```

```python
def run_scored(batsman_name):
  vk = delivery[delivery['batsman'] == batsman_name]
  return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).index[0]
```

```python
run_scored('CH Gayle')
```

```python
# problem is : Find the most destructive death over batsman in the history of IPL
# On the basis of strike rate = (Number of runs/Number of balls)/100
# min batsman 200 balls in over 16-20


mask = delivery['over']>15
death_overs = delivery[mask]
death_overs
```

```python

# ball kisne kitni kheli h
death_overs.groupby('batsman')['batsman_runs'].count()  
```

```python
# i need that player who play 200 balls and more
all_batsman = death_overs.groupby('batsman')['batsman_runs'].count()
```

```python
x = all_batsman>200
all_batsman[x]
```

```python
# I need only name
batsman_list = all_batsman[x].index.tolist()
```

```python
# Runs scored by all these 43 batsman
# Balls played by these 43 batsman

final = delivery[delivery['batsman'].isin(batsman_list)]
```

```python
final.groupby('batsman')['batsman_runs'].count()
```

```python
runs = final.groupby('batsman')['batsman_runs'].sum()
```

```python
balls = final.groupby('batsman')['batsman_runs'].count()
```

```python
sr = (runs/balls)*100
```

```python
sr
```
